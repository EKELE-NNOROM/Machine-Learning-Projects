```python
import pandas as pd
from pandas.tools.plotting import scatter_matrix
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.head()
```

```python
recent_grads.dropna(inplace=True)
recent_grads.shape[0]
```

```python
recent_grads.shape
```

```python
scatter_matrix(recent_grads[["ShareWomen","Unemployment_rate"]])
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
a1 = fig.add_subplot(2,2,1)
a2 = fig.add_subplot(2,2,2)
a3 = fig.add_subplot(2,2,3)
a4 = fig.add_subplot(2,2,4)

a1.hist(recent_grads['ShareWomen'])
a2.scatter(recent_grads['Unemployment_rate'],recent_grads['ShareWomen'])
a3.scatter(recent_grads['ShareWomen'],recent_grads['Unemployment_rate'])
a4.hist(recent_grads['Unemployment_rate'])

a1.get_xaxis().set_visible(False)
a2.get_xaxis().set_visible(False)
a2.get_yaxis().set_visible(False)
a4.get_yaxis().set_visible(False)

a1.set_ylabel("ShareWomen")
a3.set_xlabel("ShareWomen")
a3.set_ylabel("Unemployment_rate")
a4.set_xlabel("Unemployment_rate")

a1.set_xlim(0,1.0)
a2.set_xlim(0,0.2)
a3.set_xlim(0,1.0)
a3.set_ylim(0,0.2)
a4.set_xlim(0,0.2)

a1.set_yticklabels([0.0,0.5,1.0,1.5,2.0,2.5,3.0])
a3.set_xticklabels([0.0,0.2,0.4,0.6,0.8], rotation=90)
a3.set_yticklabels([0.00,0.05,0.10,0.15])
a4.set_xticklabels([0.0,0.05,0.10,0.15,0.20], rotation=90)

fig.subplots_adjust(wspace=0,hspace=0)
plt.show()
```

```python
recent_grads["ShareMen"] = recent_grads["Men"] / recent_grads["Total"]
arts = recent_grads[recent_grads['Major_category'] == 'Arts']
fig = plt.figure(figsize=(8,8))
a1 = fig.add_subplot(1,1,1)
locs = np.arange(0,len(arts['Major']))
width = 0.35
offset_locs = locs + width
bar_1 = a1.bar(left=locs, height=arts['ShareMen'].tolist(), width=width)
bar_2 = a1.bar(left=offset_locs, height=arts['ShareWomen'].tolist(), width=width, color='green')
a1.set_xticklabels(arts['Major'].tolist(), rotation=90)
a1.set_xticks(offset_locs)
plt.legend((bar_1,bar_2), ("ShareMen", "ShareWomen"), loc = 'upper left')
plt.grid(False)
plt.show()
```
