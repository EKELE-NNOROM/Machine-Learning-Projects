## Create a list which contains the length of the position titles

```python
length = []
for element in white_house['Position Title']:
    length.append(len(element))
```

```python
length
```

```python
#df['protected'] = pd.Series(['no', 'no', 'no', 'yes'])
white_house['Length'] = pd.Series(length)
```

```python
white_house.drop('Length', axis=1, inplace=True)
#df.drop('column_name', axis=1, inplace=True)
```

### Create a column called Length of Position which contains the length array above

```python
white_house['Length of Position'] = pd.Series(length)
```
