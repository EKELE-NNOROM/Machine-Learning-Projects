# Read in the data

```python
import pandas
import numpy
import re
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from mpl_toolkits.basemap import Basemap

data_files = [
    "ap_2010.csv",
    "class_size.csv",
    "demographics.csv",
    "graduation.csv",
    "hs_directory.csv",
    "sat_results.csv"
]

data = {}

for f in data_files:
    d = pandas.read_csv("schools/{0}".format(f))
    data[f.replace(".csv", "")] = d
```

# Read in the surveys

```python
all_survey = pandas.read_csv("schools/survey_all.txt", delimiter="\t", encoding='windows-1252')
d75_survey = pandas.read_csv("schools/survey_d75.txt", delimiter="\t", encoding='windows-1252')
survey = pandas.concat([all_survey, d75_survey], axis=0)

survey["DBN"] = survey["dbn"]

survey_fields = [
    "DBN", 
    "rr_s", 
    "rr_t", 
    "rr_p", 
    "N_s", 
    "N_t", 
    "N_p", 
    "saf_p_11", 
    "com_p_11", 
    "eng_p_11", 
    "aca_p_11", 
    "saf_t_11", 
    "com_t_11", 
    "eng_t_10", 
    "aca_t_11", 
    "saf_s_11", 
    "com_s_11", 
    "eng_s_11", 
    "aca_s_11", 
    "saf_tot_11", 
    "com_tot_11", 
    "eng_tot_11", 
    "aca_tot_11",
]
survey = survey.loc[:,survey_fields]
data["survey"] = survey
survey.head()
```

# Add DBN columns

```python
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]

def pad_csd(num):
    string_representation = str(num)
    if len(string_representation) > 1:
        return string_representation
    else:
        return "0" + string_representation
    
data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(pad_csd)
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]
data.keys()
```

# Convert columns to numeric

```python
cols = ['SAT Math Avg. Score', 'SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']
for c in cols:
    data["sat_results"][c] = pandas.to_numeric(data["sat_results"][c], errors="coerce")

data['sat_results']['sat_score'] = data['sat_results'][cols[0]] + data['sat_results'][cols[1]] + data['sat_results'][cols[2]]

def find_lat(loc):
    coords = re.findall("\(.+, .+\)", loc)
    lat = coords[0].split(",")[0].replace("(", "")
    return lat

def find_lon(loc):
    coords = re.findall("\(.+, .+\)", loc)
    lon = coords[0].split(",")[1].replace(")", "").strip()
    return lon

data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(find_lat)
data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(find_lon)

data["hs_directory"]["lat"] = pandas.to_numeric(data["hs_directory"]["lat"], errors="coerce")
data["hs_directory"]["lon"] = pandas.to_numeric(data["hs_directory"]["lon"], errors="coerce")
data["hs_directory"].head()
```

# Condense datasets

```python
class_size = data["class_size"]
class_size = class_size[class_size["GRADE "] == "09-12"]
class_size = class_size[class_size["PROGRAM TYPE"] == "GEN ED"]

class_size = class_size.groupby("DBN").agg(numpy.mean)
class_size.reset_index(inplace=True)
data["class_size"] = class_size

data["demographics"] = data["demographics"][data["demographics"]["schoolyear"] == 20112012]

data["graduation"] = data["graduation"][data["graduation"]["Cohort"] == "2006"]
data["graduation"] = data["graduation"][data["graduation"]["Demographic"] == "Total Cohort"]
data["class_size"].head()
```

# Convert AP scores to numeric

```python
cols = ['AP Test Takers ', 'Total Exams Taken', 'Number of Exams with scores 3 4 or 5']

for col in cols:
    data["ap_2010"][col] = pandas.to_numeric(data["ap_2010"][col], errors="coerce")
```

# Combine the datasets

```python
combined = data["sat_results"]

combined = combined.merge(data["ap_2010"], on="DBN", how="left")
combined = combined.merge(data["graduation"], on="DBN", how="left")

to_merge = ["class_size", "demographics", "survey", "hs_directory"]

for m in to_merge:
    combined = combined.merge(data[m], on="DBN", how="inner")

combined = combined.fillna(combined.mean())
combined = combined.fillna(0)
```

# Add a school district column for mapping

```python
def get_first_two_chars(dbn):
    return dbn[0:2]

combined["school_dist"] = combined["DBN"].apply(get_first_two_chars)
```

# Find correlations

```python
correlations = combined.corr()
correlations = correlations["sat_score"]
pandas.DataFrame(correlations.sort_values())
```

```python
correlations[survey_fields].plot(kind='barh', figsize=(7,5.5), title="SAT Score Correlations")
```

```python
correlations[survey_fields].sort_values().plot(kind='barh', figsize=(7, 5.5), title="SAT Score Correlations, Ordered")
```

**Going by the provided NYC schools survey data dictionary [here](https://data.cityofnewyork.us/Education/NYC-School-Survey-2011/mnz3-dyi8)**

The *N_s, N_t, N_p* (Number of student, teacher and parent respondents, respectively) group has quite high correlation with $satscore$. This is expected because those measures are highly correlated with total enrollment (*total_students*), which itself is highly positively correlated with SAT score. 

More interesting are the correlations between SAT score and the various student, parent and teacher response rates (*rr_s, rr_p, rr_t*). On one hand, highly engaged students would also lead to higher engagement and investment in education, resulting in higher SAT scores. However, the strong positive correlation doesn't hold for parent and teacher response rate -- in fact, the correlation is *negative* for $rr_t$. 

Safety scores (student, teacher and parent perception of safety) (*saf_p_11, saf_t_11, saf_s_11, saf_tot_11*) all correlate positively with SAT scores. This is understandable, as an unsafe learning environment would make it harder to both teach and study. 

Interestingly, the academic perception metrics (*aca_s_11, aca_p_11, aca_t_11, aca_tot_11*) also vary widely in their correlation with SAT Scores. While student perception (*aca_s_11*) is strongly positively correlated with scores, this doesn't hold true for parents and teachers. 

The parent perception of communication (*com_p_11*) is also interesting, since there is a very strong *negative* correlation between it and related SAT scores. 

# Investigating Safety Scores

```python
combined.plot.scatter(x='saf_s_11', y='sat_score', title="Student Perception of Safety vs. SAT Score")
plt.plot(numpy.arange(4,11), [1500]*len(numpy.arange(4,11)), c='green')
plt.axis([4, 10, 800, 2200])
plt.show()
```

There is a positive correction between student perception of safety and SAT scores, but it is not terribly strong. There are a handful of schools each with high safety and SAT scores, and low satefy and SAT scores, but most are clustered at a middling safety score and an SAT score slightly below the average (1500). 

Indeed, it looks like no school with a safety score below about 6.5 had a resultant SAT score above that [1500 national average](http://www.princetonreview.com/college-advice/good-sat-score-act-score). 

```python
district_avgs = combined.groupby('school_dist').aggregate(numpy.mean)
district_avgs.reset_index(inplace=True)

m = Basemap(
    projection='merc',
    resolution='l',
    llcrnrlat=40.496044, 
    urcrnrlat=40.915256, 
    llcrnrlon=-74.255735, 
    urcrnrlon=-73.700272)
m.drawmapboundary(fill_color='#85A6D9')
m.drawcoastlines(color='#6D5F47', linewidth=.4)
m.drawrivers(color='#6D5F47', linewidth=.4)
m.fillcontinents(color='#FFFFFF', lake_color='#85A6D9')

lons = district_avgs['lon'].tolist()
lats = district_avgs['lat'].tolist()
m.scatter(lons, lats, latlon=True, s=50, zorder=2, cmap='summer',
         c=district_avgs['saf_s_11'])
plt.title("Mapping of School District Avg. Student Safety Perception")
plt.show()
```

Based on [New York City boroughs](http://www.nycgo.com/boroughs-neighborhoods), it looks like Manhattan, the lower Bronx, west Queens and south Brooklyn have higher safety scores, while most of Brooklyn and east Queens have lower safety scores.

## Race and SAT Scores

```python
race_fields = ['white_per', 'asian_per', 'black_per', 'hispanic_per']
race_corr = combined.corr()['sat_score'][race_fields].plot.barh(title="Race Percentages Corr. SAT Score")
```

It looks like there is strong racial disparity when it comes to SAT scores, with SAT scores having strong positive correlations for percentages of white or asian students, and strong negative correlations for percentages of black or hispanic students. Contextually, this may be due to a high observed correlation between lack of funding for certain schools or districts, and a corresponding high number of black or hispanic students who live in the serviced area. 

```python
combined.plot.scatter(x='hispanic_per', y='sat_score', title="Percent Student Body Hispanic vs. SAT Score")
```

Interestingly, there are quite a few schools with exclusively (100%) hispanic student bodies, all of which have low SAT scores. Those schools are:

```python
pandas.DataFrame(combined[combined["hispanic_per"] > 99]["SCHOOL NAME"])
```

The schools in the output above are focused on providing education and English language instruction to recent school-age immigrants to the US. With a high percentage of students who are ESL, this could explain the related low SAT scores.

Conversely, there are also several schools with very high SAT scores and very low percentages of hispanic students. I'd like to investigate exactly why these schools have such high outlier SAT scores. They are:

```python
pandas.DataFrame(combined[(combined['hispanic_per'] < 10) & (combined['sat_score'] > 1800)]['SCHOOL NAME'])
```

The above schools are specialized STEM "magnet" schools that recieve special city funding, and require students to go through an admissions process/exam. This explains the high outlier SAT scores, as these schools are not drawing from a specific district, but are a collection of high-scoring students from all over New York City.

```python
correlations[['male_per', 'female_per']].plot.barh(title="SAT Score Corr. Gender Percentages")
```

It looks like there is a weakly positive correlation between high SAT scores and high percentages of female students, and a weakly negative correlation for high percentages of male students.

```python
combined.plot.scatter(x='female_per', y='sat_score', title="SAT Scores vs. Percentage Student Body Female")
```

Based on the scatterplot, there doesn't look to be a strong correlation between SAT scores and high percentages of female students. However, there are a cluster of schools with high SAT scores and relatively high female percentages (60-80%). 

```python
pandas.DataFrame(combined[(combined['female_per'] > 60) & (combined['sat_score'] > 1700)][['SCHOOL NAME', 'sat_score']])
```

All of the schools in that subset (high SAT scores, disproportionate levels of female students) are selective liberal arts "magnet" schools in NYC that employ admissions extrance exams.


Notably, there are also a handful of 100% female, low-SAT schools. 

```python
pandas.DataFrame(combined[(combined['female_per'] > 95) & (combined['sat_score'] < 1400)][['SCHOOL NAME', "sat_score"]])
```

It looks like these are all girls-only schools within NYC that focus on female student academics, but are also part of the public NYC-school district entrance lottery (therefore, no admissions process and entrance exam). 

## AP Scores and SAT Scores

```python
combined['ap_per'] = combined['AP Test Takers '] / combined['total_enrollment']
combined.plot.scatter(x='ap_per', y='sat_score', title="SAT Score vs. Percentage Student Body AP Test-Takers")
```

It looks like there is a weakly positive correlation between percentage of AP test takers in a school, and average SAT score. Interesting to note, however, is the line of SAT scores at 1200 that continues horizontally regardless of increases in percentage of AP test takers. Perhaps these are schools where AP testing is almost to entirely mandatory?
