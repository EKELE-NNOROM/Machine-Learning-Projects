```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
%matplotlib inline
```

```python
movies = pd.read_csv("fandango_score_comparison.csv")
movies
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2,figsize = (10,4))
p1 = movies['Fandango_Stars'].plot.hist(ax=axs[0],title='Fandango')
p2 = movies['Metacritic_norm_round'].plot.hist(ax=axs[1],title='Metacritic')
p1.set_xlim(0,5.0)
p2.set_xlim(0,5.0)
```

```python
fig,axs = plt.subplots(nrows=1, ncols=2,figsize=(10,4))
p1 = movies['Fandango_Stars'].plot.hist(ax=axs[0],title='Fandango')
p2 = movies['Metacritic_norm_round'].plot.hist(ax=axs[1],title='Metacritic')
p1.set_xlim(0,5.0)
p2.set_xlim(0,5.0)
```

```python
fg_score = movies['Fandango_Stars']
mc_score = movies['Metacritic_norm_round']
print(fg_score.mean(), mc_score.mean())
print(fg_score.median(), mc_score.median())
print(fg_score.std(), mc_score.std())
```

```python
movies.plot.scatter(x='Fandango_Stars',y='Metacritic_norm_round')
plt.grid(True)
```

```python
movies['fm_diff'] = mc_score - fg_score
```

```python
movies.head()
```

```python
movies_by_fm_diff = movies.sort_values('fm_diff',ascending=False)
```

```python
movies_by_fm_diff
```

```python
movies['fm_diff'] = np.abs(mc_score - fg_score)
movies_by_fm_diff = movies.sort_values('fm_diff',ascending=False)
```

```python
movies_by_fm_diff
```

```python
fm_r, fm_p = sp.stats.pearsonr(fg_score,mc_score)
fm_r, fm_p
```

### Seeing the very low correlation we cannot use Metacritic values to determine Fandango values
#### That is, as metacritic values increase there is a lower likelihood of a relationship to Fandango values

```python
fm_linreg = sp.stats.linregress(mc_score,fg_score)
fm_linreg
```

```python
f_3 = fm_linreg.slope * 3.0 + fm_linreg.intercept
```

```python
f_3
```

```python
# A movie with a 4.0 from Metacritic is predicted to get from Fandango:

f_4 = fm_linreg.slope * 4.0 + fm_linreg.intercept
f_4
```

```python
f_1 = fm_linreg.slope * 1.0 + fm_linreg.intercept
f_2 = fm_linreg.slope * 2.0 + fm_linreg.intercept
f_0 = fm_linreg.slope * 0.0 + fm_linreg.intercept
```

```python
plt.scatter([3,4],[f_3,f_4])
plt.grid(True)
```

```python
movies['rt_diff'] = rt_diff
```

```python
movies_by_rt_diff = movies.sort_values('rt_diff',ascending=False)
movies_by_rt_diff
```

### So Do You Believe, Little Boy and Hitman:Agent 47 have the highest difference in values between the users and critics in Rotten Tomato

### CORRELATION

### The user and critic values for rotten tomato have a strong correlation
