```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
%matplotlib inline
```

```python
movies = pd.read_csv("fandango_score_comparison.csv")
movies
```

### Fandango vs. Metacritic

```python
fig, axs = plt.subplots(ncols=2, figsize=(10,4))
p1 = movies["Fandango_Stars"].plot.hist(ax=axs[0], title='Fandango')
p2 = movies['Metacritic_norm_round'].plot.hist(ax=axs[1], title='Metacritic')
p1.set_xlim(0, 5.0)
p2.set_xlim(0, 5.0)
```

It looks like the Fandango scores are much more closely grouped (low variance), all above 3/5 stars. The Metacritic review spread is comparatively more "normally" distributed over the 0-5.0 normalized rating scale. 

```python
fg_score = movies['Fandango_Stars']
mc_score = movies['Metacritic_norm_round']
print(fg_score.mean(), mc_score.mean())
print(fg_score.median(), mc_score.median())
print(fg_score.std(), mc_score.std())
```

```python
movies['fm_diff'] = np.abs(mc_score - fg_score)
movies_by_fm_diff = movies.sort_values("fm_diff", ascending=False)
movies_by_fm_diff.head(5)
```

### Looking at Correlation

```python
# calculating r-value (pearson)
fm_r, fm_p = sp.stats.pearsonr(fg_score, mc_score)
fm_r, fm_p # r-value, p-value for non-correlation
```

The correlation is actually fairly low. Write up a markdown cell that discusses what this might mean.

```python
fm_linreg = sp.stats.linregress(mc_score, fg_score)
fm_linreg
```

```python
# A movie with a 3.0 from Metacritic is predicted to get from Fandango:

f_3 = fm_linreg.slope * 3.0 + fm_linreg.intercept
f_3
```

```python
# A movie with a 4.0 from Metacritic is predicted to get from Fandango:

f_4 = fm_linreg.slope * 4.0 + fm_linreg.intercept
f_4
```

```python
plt.scatter(x=[3.0, 4.0], y=[f_3, f_4])
plt.show()
```
